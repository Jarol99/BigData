```python
##Cuadrado dos dimensiones

from keras.layers import Input,Dense,Conv2D,MaxPooling2D,UpSampling2D, Flatten
from keras.models import Model
import matplotlib.pyplot as plt
import numpy as np
import cv2


def reformar_filtros(filtros):
  l=len(filtros)
  m=len(filtros[0])
  n=len(filtros[0][0])
  salida=np.zeros((m,n,1,l))
  print('arreglo',l,n,m,salida.shape)
  for i in range(l):
    for j in range(m):
      for k in range(n):
        #print('ijk',i,j,k,filtros[i][j][k])
        salida[j,k,0,i]=filtros[i][j][k]
  return salida

alto = 50;
ancho = 200;

mi_arreglo = np.zeros((alto,ancho))

mi_arreglo[10:40,40:150] = 1;

plt.imshow(-mi_arreglo,cmap='gray')

#####################################



filtro1=[
    [1, -1],
    [1, -1]
]
filtro2=[
    [1, 1],
    [-1, -1]
]


filtros=reformar_filtros([filtro1,filtro2])

input_img = Input(shape=(50,200,1))
x = Conv2D(2,(2,2),activation='linear',padding='same')(input_img)
modelo3 = Model(input_img,x)

modelo3.set_weights([filtros,np.array([0,0], dtype=np.float32)])

img_rz=np.reshape(mi_arreglo,(1,50,200,1))
img_tmp3=modelo3.predict(img_rz)

img_fin1 = np.reshape(img_tmp3[0,:,:,0],(50,200))
plt.imshow(img_fin1,cmap='gray')
plt.show()
img_fin2 = np.reshape(img_tmp3[0,:,:,1],(50,200))
plt.imshow(img_fin2,cmap='gray')
plt.show()
```

```python
##Cuadrado tres dimensiones



from keras.layers import Input,Dense,Conv2D,MaxPooling2D,UpSampling2D, Flatten
from keras.models import Model
import matplotlib.pyplot as plt
import numpy as np
import cv2


def reformar_filtros(filtros):
  l=len(filtros)
  m=len(filtros[0])
  n=len(filtros[0][0])
  salida=np.zeros((m,n,1,l))
  print('arreglo',l,n,m,salida.shape)
  for i in range(l):
    for j in range(m):
      for k in range(n):
        #print('ijk',i,j,k,filtros[i][j][k])
        salida[j,k,0,i]=filtros[i][j][k]
  return salida

alto = 50;
ancho = 200;

mi_arreglo = np.zeros((alto,ancho))

mi_arreglo[10:40,40:150] = 1;

plt.imshow(-mi_arreglo,cmap='gray')

#####################################



filtro1=[
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
]
filtro2=[
    [1, 1, 1],
    [0, 0, 0],
    [-1, -1, -1]
]

filtro3=[
    [1, 2, 1],
    [2, -12, 2],
    [1, 2, 1]
]

filtros=reformar_filtros([filtro1,filtro2,filtro3])

input_img = Input(shape=(50,200,1))
x = Conv2D(3,(3,3),activation='linear',padding='same')(input_img)
modelo3 = Model(input_img,x)

modelo3.set_weights([filtros,np.array([0,0,0], dtype=np.float32)])

img_rz=np.reshape(mi_arreglo,(1,50,200,1))
img_tmp3=modelo3.predict(img_rz)

img_fin1 = np.reshape(img_tmp3[0,:,:,0],(50,200))
plt.imshow(img_fin1,cmap='gray')
plt.show()
img_fin2 = np.reshape(img_tmp3[0,:,:,1],(50,200))
plt.imshow(img_fin2,cmap='gray')
plt.show()
img_fin3 = np.reshape(img_tmp3[0,:,:,2],(50,200))
plt.imshow(img_fin3,cmap='gray')

```
